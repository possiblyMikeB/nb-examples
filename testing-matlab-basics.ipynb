## Basic Info & Ploting
The contents of a code cell are evaluated using an instance of MATLAB runing in the background; as such you have full access to the language and most installed libraries. In addition, once you've specified an output format for plots, the plot functions behave as expected.

```matlab
% simple pair of vectors discretizing a function
x=linspace(0,1,100);
y=sin(x).^2;

% sets plot output type for *Entire Notebook* 
imatlab_export_fig('print-png')

% plot of their relationship and something else.
hold on
plot(x,y)
plot(x.^3,y)
```
